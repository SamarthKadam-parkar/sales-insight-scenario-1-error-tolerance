# Project Setup Notebook

```
USE DATABASE SALES_DB;

CREATE SCHEMA IF NOT EXISTS SALES_DB.RAW;
CREATE SCHEMA IF NOT EXISTS SALES_DB.SILVER;

CREATE STAGE IF NOT EXISTS SALES_DB.PUBLIC.SALES_STAGE
ENCRYPTION = (TYPE = 'SNOWFLAKE_SSE');

CREATE FILE FORMAT IF NOT EXISTS SALES_DB.PUBLIC.CSV_FF
TYPE = 'CSV'
SKIP_HEADER = 1 
FIELD_DELIMITER = ','
```

```
LIST @SALES_STAGE;
```

```
CREATE TABLE SALES_DB.RAW.PRODUCTS AS 
(
    SELECT 
         $1 AS PRODUCT_ID
        ,$2 AS PRODUCT_NAME
        ,$3 AS PRODUCT_CATEGORY
        ,$4 AS PRODUCT_COST
    FROM @SALES_DB.PUBLIC.SALES_STAGE/Products/products.csv
    (FILE_FORMAT => 'CSV_FF')
);
```

```
CREATE TABLE SALES_DB.RAW.CUSTOMERS AS
(   
    SELECT 
         $1 AS CUSTOMER_ID
        ,$2 AS CUSTOMER_NAME
        ,$3 AS CUSTOMER_CITY
        ,$4 AS CUSTOMER_COUNTRY
    FROM @SALES_DB.PUBLIC.sales_stage/Customers/customers.csv
    (FILE_FORMAT => 'CSV_FF')
);
```

```
CREATE TABLE SALES_DB.RAW.STORES AS
(   
    SELECT 
         $1 AS STORE_ID
        ,$2 AS STORE_NAME
        ,$3 AS STORE_CITY
        ,$4 AS STORE_COUNTRY
    FROM @SALES_DB.PUBLIC.sales_stage/Stores/stores.csv
    (FILE_FORMAT => 'CSV_FF')
);
```

```
%%sql -r dataframe_6
CREATE TABLE SALES_DB.RAW.SALES AS 
(
    SELECT 
         $1 AS SALES_ID
        ,$2 AS CUSTOMER_ID
        ,$3 AS PRODUCT_ID
        ,$4 AS STORE_ID
        ,$5 AS ORDER_DATE
        ,$6 AS QUANTITY
        ,$7 AS UNIT_PRICE
        ,$8 AS DISCOUNT
        ,$9 AS TOTAL_AMOUNT
    FROM @SALES_DB.PUBLIC.SALES_STAGE/Sales/sales.csv
    (FILE_FORMAT => 'CSV_FF')
);
```
